```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('company_sales_data.csv')
data
```

```python
sns.lineplot(x='month_number', y='total_profit', data=data)
plt.show()
```

```python
sns.lineplot(x=data["month_number"],y=data["facecream"],label="facecream")
sns.lineplot(x=data["month_number"],y=data["facewash"],label="facewash")
sns.lineplot(x=data["month_number"],y=data["toothpaste"],label="toothpaste")
sns.lineplot(x=data["month_number"],y=data["bathingsoap"],label="bathingsoap")
sns.lineplot(x=data["month_number"],y=data["shampoo"],label="shampoo")
sns.lineplot(x=data["month_number"],y=data["moisturizer"],label="moiturizer")
plt.title("Line plot")
plt.xlabel('sales')
plt.ylabel('profit')
plt.legend()
plt.show()
```

```python
sns.barplot(x=data["month_number"],y=data["facecream"],label="facecream")
sns.barplot(x=data["month_number"],y=data["facewash"],label="facewash")
sns.barplot(x=data["month_number"],y=data["toothpaste"],label="toothpaste")
sns.barplot(x=data["month_number"],y=data["bathingsoap"],label="bathingsoap")
sns.barplot(x=data["month_number"],y=data["shampoo"],label="shampoo")
sns.barplot(x=data["month_number"],y=data["moisturizer"],label="moiturizer")
plt.title("Line plot")
plt.xlabel('sales')
plt.ylabel('profit')
plt.legend()
plt.show()
```
